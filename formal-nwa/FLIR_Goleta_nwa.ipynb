```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```python

formal_nw = nx.Graph()

with open('formal_network.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    #nw_data = []
    #nw_data.append([],[],[],[])
    for row in csv_reader:
        if line_count == 0:
            print("skip")
            line_count=line_count+1
        elif line_count == 1:
            formal_nw.add_node(int(row[4]), role=row[0], name=row[1], group=row[2])
            line_count=line_count+1
        else:
            formal_nw.add_node(int(row[4]), role=row[0], name=row[1], group=row[2])
            formal_nw.add_edge(int(row[4]), int(row[3]))
            line_count=line_count+1
            
            
            
    print(f'Processed {line_count} lines.')
```

```python
#plt.subplot(111)
nx.draw_spring(formal_nw, with_labels=True, font_weight='bold')
```

```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
import pygraphviz as pgv
```

```python

formal_nw = nx.Graph()

with open('formal_network.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    #nw_data = []
    #nw_data.append([],[],[],[])
    for row in csv_reader:
        if line_count == 0:
            print("skip")
            line_count=line_count+1
        elif line_count == 1:
            formal_nw.add_node(int(row[4]), role=row[0], name=row[1], group=row[2])
            line_count=line_count+1
        else:
            formal_nw.add_node(int(row[4]), role=row[0], name=row[1], group=row[2])
            formal_nw.add_edge(int(row[4]), int(row[3]))
            line_count=line_count+1
            
            
            
    print(f'Processed {line_count} lines.')
```

```python
closeness = nx.closeness_centrality(formal_nw)
```

```python
degree = nx.degree_centrality(formal_nw)
```

```python
betweenness = nx.betweenness_centrality(formal_nw)
```

```python


plt.figure(figsize=(15,15))  # create a figure object
#mesh_ax = nw.add_subplot(1, 1, 1) 
nx.draw_spring(formal_nw, with_labels=True, node_color=list(closeness.values()))
```
